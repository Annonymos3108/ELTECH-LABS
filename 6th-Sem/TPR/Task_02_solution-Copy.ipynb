## **Задача 2**

<p style='text-align: justify;'>Технология краудсорсинга позволяет привлекать широкий круг людей к задачам сбора и обработки информации. Заказчики могут размещать на платформе краудсорсинга задания, назначая за них определенное вознаграждение, а пользователи платформы (исполнители) выполняют эти задания и получают назначенное вознаграждение. Примерами таких платформ являются Amazon Mechanical Turk, Яндекс.Толока. Одной из разновидностей краудсорсинга является пространственный краудсорсинг, при котором задания имеют пространственную привязку (например, с помощью системы пространственного краудсорсинга можно сделать фотографию определенной географической локации в определенный момент).</p>

<p style='text-align: justify;'>В данной задаче рассматривается один из центральных компонентов платформы пространственного краудсорсинга, осуществляющий распределение заданий среди исполнителей.</p>

<p style='text-align: justify;'>Задано определенное множество заданий (см. табл. 1). Каждое задание имеет пространственную привязку. Одной из функций подобной привязки являтся идентификация исполнителей, находящихся рядом с заданием. Для простоты будем полагать, что задания и исполнители находятся в городе Гипподамовой системы, что позволяет использовать локальную систему координат и оценивать расстояние с помощью манхеттенской метрики. Задания считаются достаточно простыми, чтобы любое из них могло быть выполнено любым участником.</p>

<p style='text-align: justify;'>В заданный момент в системе активно определенное множество участников (табл. 2). Для каждого участника задано текущее положение в локальной системе координат, а также текущий рейтинг (в диапазоне от 0 до 1 - чем больше, тем более надежным является исполнитель). </p>

<p style='text-align: justify;'>Вознаграждение исполнителя определяется его рейтингом. В частности, задано два граничных уровня рейтинга $ Th_M = 0.3 $ (средний) и $ Th_H = 0.7 $ (высокий). Участники с высоким рейтингом ($ r_i \geq Th_H $) получают вознаграждение 90 денежных единиц (ДЕ) за каждое задание, участники со средним рейтингом ($ Th_M \leq ri < Th_H $) получают вознаграждение 60 ДЕ. Участники с рейтингом ниже $ Th_M $ считаются заблокированными и к выполнению заданий не привлекаются.
Бюджет заказчика на выполнение заданного набора заданий составляет $ 300 $ ДЕ. </p>


Требуется:    
1. <p style='text-align: justify;'>Считая качество выполнения задания численно равным рейтингу исполнителя, найти распределение (всех) заданий по исполнителям, максимизирующее суммарное качество выполнения заданий.</p>

### Некоторые рассуждения

---

<p style='text-align: justify;'>Задача распределения задач пространственного краудсорсинга с бюджетными ограничениями сформулирована как задача о рюкзаке с множественным выбором</p>

<p style='text-align: justify;'>Мы рассматриваем проблему делегирования задачи пространственного краудсорсинга $\tau_i$, предложенной заявителем $i$, $N$ кандидатам в рабочие с ограничением бюджета $B^{\tau_i}\in\mathbb{R}^+$.</p>

<p style='text-align: justify;'>Пространственная задача $\tau_i$ представлена в виде кортежа вида $ \left \langle l^{\tau_i},R^{\tau_i},B^{\tau_i},p_H^{\tau_i},p_M^{\tau_i} \right \rangle $. Информация, относящаяся к пространственной задаче, включает в себя $l^{\tau_i}$, который является местоположением задачи, заданной точкой в 2D-пространстве, и $R^{\tau_i}$, который представляет собой радиус пространственной области, в которой исполнители с наибольшей вероятностью примут поставленную задачу.</p>

Дано:
    
$$ Th_{HM} = 0.7 ~~~~~~~~~~~~~~~~~~~~~~~~(1)\\ 
Th_{ML} = 0.3  ~~~~~~~~~~~~~~~~~~~~~~~~~(2)\\ 
0\le Th_{ML}\le Th_{HM}\le1 ~~~~~(3) \\
p_H^{\tau_i} = 90 ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(4)\\
p_M^{\tau_i} = 60 ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(5)\\
B^{\tau_i} = 300 ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~(6)\\
$$

<p style='text-align: justify;'>Kачество результата зависит от двух основных факторов:</p>

1. <p style='text-align: justify;'>На него влияет надежность исполнителя. Более заслуживающий доверия исполнитель (о чем свидетельствует его хорошая репутация) обычно обеспечивает более надежные результаты, чем менее заслуживающие доверия исполнители.</p>
2. <p style='text-align: justify;'>Расстояние между местонахождением исполнителя и местом выполнения пространственной задачи влияет на стоимость проезда исполнителея (и, следовательно, на его готовность поехать к месту выполнения задачи), что, в свою очередь, влияет на вероятность того, что запрашивающий получит качественные результаты от этого исполнителя.</p>

<p style='text-align: justify;'>Например, достоверность результата, предоставленного исполнителем $1$, находящимся на расстоянии $d_1$ от местоположения задачи, должна быть уменьшена больше, чем достоверность результата, предоставленного исполнителем $2$, находящимся на расстоянии $d_2$ от местоположения задачи, для $d_1>d_2$. Основываясь на совокупном рассмотрении этих двух факторов, доверие к работнику $j$ в контексте пространственной задачи $ \tau_i $ составляет:</p>

$$ c_j^{\tau_i}=r_j\cdot\delta\left(l_j,l^{\tau_i}\right) ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ (7)$$ 
    

<p style='text-align: justify;'>где $l_j$ - местонахождение работника в момент делегирования ему запроса задачи. $\delta\left(l_j,l^{\tau_i}\right)$-  функция, вычисляющая скидку на репутацию работника в результате его близости к месту выполнения задачи. Это определяется как:</p>

$$ \delta\left(l_j,l^{\tau_i}\right)=1-max\left[0,min\left[{log}_{D_c}\left(d\left(l_j,l^{\tau_i}\right)\right),1\right]\right] ~~~~~~~~~~ (8)$$ 

<p style='text-align: justify;'>где $d\left(l_j,l^{\tau_i}\right)$ вычисляет евклидово расстояние между двумя координатами.</p>
<p style='text-align: justify;'>Предполагаем, что максимальный диапазон поиска кандидатов в исполнители находится в пределах того же города, что и предлагаемая пространственная задача. Скидка на расстояние является функцией ${log}_{D_c}$, где $D_c$- диаметр данной области $c$ (города). Следовательно, $\delta\left(l_j,l^{\tau_i}\right)\in[0\ ,1]$. Чем ближе рабочий $j$ к месту выполнения задания $\tau_i$, тем ближе $\delta\left(l_j,l^{\tau_i}\right)$ к $1$. Когда рабочий $j$ находится на расстоянии $D_c$ от $\tau_i$ или дальше, $\delta\left(l_j,l^{\tau_i}\right)=0$.</p>

```python
from cvxopt import matrix, solvers

def solve():
    c = matrix([-0., -60., -60., -90., -60., -60., -90., -60., -60., -60.])       
    G = matrix([[  0,   60, 60, 90, 60, 60, 90, 60, 60, 60],
                [  0,   60, 60, 90, 60, 60, 90, 60, 60, 60],
                [  0,   60, 60, 90, 60, 60, 90, 60, 60, 60],
                [  0,   60, 60, 90, 60, 60, 90, 60, 60, 60]])
    h = matrix([300, 300, 300, 300], tc='d')
    solution = solvers.lp(c, G.T, h, solver='glpk')
    return (solution['status'] == 'optimal', -solution['primal objective']-55000, solution['x'], solution)

# Решение задачи со средними значениями параметров
solution = solve()

print('Status:', solution[0])
print('Objective:', solution[1])
print('x = \n', solution[2])
```

### Формальная постановка (Пункт 2)

Максимизировать:

$$ \sum_{j=1}^{N}c_j^{\tau_i} ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ (9)$$ 

*при условии:*

$$ \sum_{j=1}^{N}\sum_{g=1}^{G}{\psi_{j,g}p_{j,g}\le B^{\tau_i}} ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ (10) \\
\psi_{j,g}\in{0,1}; ~~ \forall j\in\ {1,\ ...,\ N},\ \forall g\in\ {1,\ ...,\ G} ~~~~~~~ (11) \\
\sum_{g=1}^{G}{\psi_{j,g}=1;\ \forall j\ \in\ {1,\ ...,\ N}} ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ (12) $$ 


<p style='text-align: justify;'>где $p_{j,g}\in{p_H^{\tau_i},p_M^{\tau_i},0}$ представляет собой фактическое вознаграждение работника $j$. $G$ обозначает количество вариантов вознаграждения за каждую задачу ($G = 3$ в этой работе). Значение $\psi_{j,g}$ зависит от репутации $j$. Таким образом, если $r_j\geq Th_{HM}$, то только $\psi_{j,1}  = 1$ (т.e. $j$ выплачивается $p_H^{\tau_i}$ ДЕ, если запрашивающий принимает его результат); если $Th_{ML} \le r_j \le Th_{HM}$, то только $\psi_{j,2}=1$ (т.е. $j$ выплачивается $p_M^{\tau_i}$ ДЕ, если запрашивающий принимает его результат); в противном случае только $\psi_{j,3}=1$ (т.е. $j$ получает $0$ ДЕ, что означает, что $j$ не приглашен для участия в задаче $\tau_i$)</p>

**Запишем формулы (13-16) в развернутом виде**

Максимизировать:

$$ \sum_{j=1}^{10}c_j^{\tau_i} \rightarrow max ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ (13)$$ 

*при условии:*

$$ \sum_{j=1}^{10}\sum_{g=1}^{3}{\psi_{j,g}p_{j,g}\le 300} ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ (14) \\
\psi_{j,g}\in{0,1}; \forall j\in {1,10}, \forall g\in\ {1,3} ~~~~~~~~~~~~~~~~~~~~~~~~~~~~ (15) \\
\sum_{g=1}^{3}{\psi_{j,g}=1; \forall j \in {1,10}} ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ (16) $$ 

```python
# Task_02_2
from math import log, floor
from cvxopt import matrix, solvers
from scipy.spatial import distance
from numpy import array, zeros, argsort

# Исходные данные
N = 10
T = 4
G = 3
Th_HM   = 0.7
Th_ML   = 0.3
p_ti_H  = 90
p_ti_M  = 60
B_ti    = 300
D_c     = 19   # Диаметр данной области
R_ti    = 27   # Радиус (макс возможный)


tasks   = list([[ 1, [1, 6]],
                [ 2, [11,8]],
                [ 3, [5,17]],
                [ 4, [7,6]]])
workers = list([[ 1,   [8,4], 0.25],
                [ 2,   [4,1], 0.32],
                [ 3, [12,14], 0.47],
                [ 4, [11,14], 0.88],
                [ 5,   [5,1],  0.4],
                [ 6, [13,19], 0.59],
                [ 7,  [5,19], 0.75],
                [ 8,  [3,19], 0.61],
                [ 9, [16,13], 0.49],
                [10,  [6,17], 0.57]])


def reputation_discount(worker_location, task_location):
    return 1 - max(0, min(log(distance.euclidean(worker_location,task_location), D_c),1))

def solve(task_i):    
    confidence_c_ti_j = zeros(shape=(N,2))
    C_ti_S_j_collection = []
    for j in range(0,N):
        confidence_c_ti_j[j][0] = j + 1
        #formula (7)
        confidence_c_ti_j[j][1] = workers[j][2]*reputation_discount(array(workers[j][1]), array(tasks[task_i][1])) 
    
    # Сортировка полученных репутаций работников по убыванию 
    sorted_confidence = (confidence_c_ti_j)[confidence_c_ti_j[:,1].argsort()[::-1]]
    
    for j in range (0, min(int(floor(B_ti/p_ti_H)), N)):
        J = min(N-j, int(floor(B_ti-j*p_ti_H/p_ti_M)))
        W_j = []
        S_j = []
        for k in range(0,N):
            if distance.euclidean(array(workers[k][1]), array(tasks[task_i][1])) <= R_ti and workers[k][2] >= Th_ML:
                W_j.append(workers[k])
            if len(W_j)>=j+J:
                break
        for k in range(0,len(W_j)):
            if workers[k][2] >= Th_HM:
                S_j.append(workers[k])
        #print(W_j)
        print(S_j)
        
        C_ti_S_j = 0.
        for k in range(0,len(S_j)):
            C_ti_S_j += sorted_confidence[k][1]
        C_ti_S_j_collection.append(C_ti_S_j)
        
    P_ti = max(C_ti_S_j_collection)
    print(C_ti_S_j_collection)  
    return P_ti

# для одной задачи 
print(solve(0))
    
```
